## 쇼핑몰 고객 주문 데이터 파악
- 데이터를 확인하여 모델 수립 혹은 목표 설정

```python
import pandas as pd
import numpy as np
```

### 데이터 셋
- 온라인 리테일 사이트의 2010/12 ~ 2011/12간 주문 기록 데이터
- 약 500,000건의 데이터
- 데이터 출처 : UCL ML Repositor.
- 데이터 세트 컬럼
    - InvoiceNo : 주문번호  
    - StockCode : 아이템 품목  
    - Description : 상품 설명  
    - Quantity : 상품 주문 수량  
    - InvoiceDate : 상품 주문 시각
    - UnitPrice : 상품 가격
    - CustomerID : 주문 고객 아이디
    - Country : 고객 거주 국가(지역)

```python
retail = pd.read_csv('OnlineRetail.csv')
retail.head()
```

### 데이터 클렌징

```python
retail.info()
```

```python
retail.isnull().sum()
```

실제 데이터에서는 CustomerID가 존재하지않는 데이터가 존재할 수 없다. 그렇기에 해당 로직에서 삭제하겠습니다.

```python
retail = retail[pd.notnull(retail['CustomerID'])]
retail
```

```python
retail.describe()
```

```python
retail = retail[retail['Quantity'] > 0]
retail = retail[retail['UnitPrice'] > 0]
len(retail)
```
